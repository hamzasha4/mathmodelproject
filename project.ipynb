```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.isotonic import IsotonicRegression
from sklearn.utils import check_random_state
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
from math import sqrt

n = 1896
x = np.arange(n)+1
df = pd.read_csv('data2.csv')
df.head()
y = df['Price']
n = y.size
x = np.arange(n)+1
```

```python
yn = []
for i in range(y.size-1,-1,-1):
    yn.append(y[i])
yn
```

```python
plt.plot(x,yn)
```

```python
ir = IsotonicRegression(out_of_bounds="clip")
y_ = ir.fit_transform(x, yn)
```

```python
plt.plot(x,yn,x,y_)
```

```python
lr = LinearRegression()
lr.fit(x[:, np.newaxis], yn)  
```

```python
ynn = lr.predict(x[:, np.newaxis])
plt.plot(x,yn,x[:, np.newaxis],ynn)
```

```python
score = accuracy_score(yn,ynn)
```

```python
y_ = ir.predict(x)
plt.plot(x,yn,x,y_)
```
